```python
for col in cat_cols:
    freqs=df_all[col].value_counts()
    k= freqs.index[freqs>20][:-1]
    for cat in k:
        name=col+'_'+cat 
        df_all[name]=(df_all[col]==cat).astype(int)
    del df_all[col]
    print(col)
```
